```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
data=pd.read_csv('uber-raw-data-apr14.csv')
```

```python
data
```

```python
data['Date/Time']=data['Date/Time'].map(pd.to_datetime)
```

```python
def get_day(dt):
    return dt.day
data['Day']=data['Date/Time'].map(get_day)

def get_hours(dt):
    return dt.hour
data['Hour']=data['Date/Time'].map(get_hours)

def get_weekday(dt):
    return dt.weekday()
data['WeekDay']=data['Date/Time'].map(get_weekday)
```

```python
data
```

```python
plt.hist(data.Day,bins=30,range=(0.5,30.5),rwidth=.8)
plt.xlabel('Date of the month')
plt.ylabel('frequency')
plt.title('Frequency by Day-uber-Apr2014')
plt.show()
```

```python
def count_rows(rows):
    return len((rows))
by_date=data.groupby('Day').apply(count_rows)
by_date
```

```python
plt.bar(range(1,31),by_date)
plt.show()
```

```python
by_date_sortvalues=by_date.sort_values()
by_date_sortvalues
```

```python
plt.bar(range(1,31),by_date_sortvalues)
plt.xticks(range(1,31),by_date_sortvalues.index)
plt.xlabel('Date of the month')
plt.ylabel('frequency')
plt.title('Frequency by Day-uber-Apr2014')
plt.show()
```

# analyze the weekday

```python
plt.hist(data.WeekDay,bins=7,range=(-.5,6.8),rwidth=0.8,color='#AA6666',alpha=.4)
plt.xticks(range(7),'Mon Tue Wed Thu Fri Sat Sun'.split())
plt.show()
```

# cross analysis(hour,Day)

```python
by_cross=data.groupby('WeekDay Hour'.split()).apply(count_rows).unstack()
by_cross
```

```python
seaborn.heatmap(by_cross)
plt.show()
```
